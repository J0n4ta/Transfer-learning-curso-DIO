#Primeira Célula
Importações de bibliotecas necessárias, como **cv2** para processamento de imagens, **numpy** para cálculos com arrays, **matplotlib** para exibir imagens e gráficos e a última sendo uma função especial para mostrar imagens no Colab.

Após isso a imagem está sendo baixada no formato png, logo depois carregada e preparada usando **cv2.imread()** que lê a imagem do arquivo, além de usar **cv2.cvColor()** que converte de BGR para RGB.

Por último, mostra a imagem original com 6x6 polegadas e depois mostrando a imagem, fora os detalhes de título, remoção dos eixos e o mais importante exibição da imagem, além de dizer as dimensões da imagem.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# Baixar a imagem Lena
!wget https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png -O lena.png

# Carregar a imagem
lena = cv2.imread('lena.png')
lena = cv2.cvtColor(lena, cv2.COLOR_BGR2RGB)  # Converter para RGB

# Mostrar a imagem original
plt.figure(figsize=(6,6))
plt.imshow(lena)
plt.title('Imagem Lena Original')
plt.axis('off')
plt.show()

print("Dimensões originais:", lena.shape)
```

#Segunda Célula
Carregando a imagem com **plt.imread()** e em seguida convertendo a imagem para tons de cinza usando **np.dot()**. Além de usar **img[..., 3:]** onde pega todos os pixels e os 3 canais de cores e depois adicionando os pesas que são **[0.299, 0.587, 0.114]**, onde são padrões para conversão RGB para cinza. Explicando melhor sobre **np.dot** no qual faz a multiplicação matricial da imagem:
**0.299 * R + 0.587 * G + 0.114 * B**

Para conversão da imagem para preto e branco, foi necessário definir um limiar, onde literalmente para deixar um imagem preto e branco tem o processo de **Binarização**, para tal processo é necessário usar **threshold** de 0.5, onde as imagens são normalizadas entre 0 e 1. Além disso, foi usado **.astype(float)** para converter o resultado booleano para números entre 0 e 1.

Em seguida, foi feito a exibição dos resultados, onde foi criado uma imagem com 15x5 polegadas e depois mostrando as 3 versões da imagem lado a lado. Já os códigos **cmap='gray'**, defini a escala de cores como preto e branco, e **axis('off')**, remove os eixos numéricas ao redor da imagem.



```python
# Carregar imagem como array numpy
img = plt.imread('lena.png')

# Converter para cinza
gray_img = np.dot(img[...,:3], [0.299, 0.587, 0.114])

# Aplicando threshold para converter em preto e branco
threshold = 0.5
bw_img = (gray_img > threshold).astype(float)

# Mostrar resultados
plt.figure(figsize=(15,5))

# Imagem original
plt.subplot(1,3,1)
plt.imshow(img)
plt.title('Original')
plt.axis('off')

# Tons de cinza
plt.subplot(1,3,2)
plt.imshow(gray_img, cmap='gray')
plt.title('Tons de Cinza')
plt.axis('off')

# Preto e branco
plt.subplot(1,3,3)
plt.imshow(bw_img, cmap='gray')
plt.title(f'Preto e Branco')
plt.axis('off')

plt.show()
```
